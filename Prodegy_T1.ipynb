```python
import pandas as pd
```

```python
# Load dataset
data = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
```

```python
data.head(5)
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
# Select latest year (example: 2022)
latest_year = "2022"

# Sort population
top_countries = data.sort_values(by=latest_year, ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(top_countries['Country Name'], top_countries[latest_year])
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(data[latest_year].dropna(), bins=20)
plt.title('Population Distribution Across Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```
