```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")   # change file name if different
data.head()
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
data.drop('Cabin', axis=1, inplace=True)
```

```python
sns.countplot(x='Survived', data=data)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title("Survival by Gender")
plt.show()
```

```python
plt.hist(data['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(numeric_data.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
#After performing data cleaning and EDA on the Titanic dataset, it was observed that gender and passenger class strongly influenced survival. Females and higher-class passengers had higher survival rates. Age distribution shows most passengers were adults. The correlation heatmap indicates fare and class are important factors affecting survival.
```
