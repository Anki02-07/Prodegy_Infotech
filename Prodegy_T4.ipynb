```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
data = pd.read_csv("twitter_training.csv")
```

```python
data.head()
```

```python
def get_sentiment(text):
    polarity = TextBlob(str(text)).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
data['Sentiment'] = data['im getting on borderlands and i will murder you all ,'].apply(get_sentiment)
data.head()
```

```python
sns.countplot(x='Sentiment', data=data)
plt.title("Sentiment Distribution")
plt.show()
```

```python
data['Sentiment'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Sentiment Percentage")
plt.ylabel("")
plt.show()
```

```python
from collections import Counter

all_words = " ".join(data['im getting on borderlands and i will murder you all ,'].astype(str))
word_counts = Counter(all_words.split())
print(word_counts.most_common(10))
```
